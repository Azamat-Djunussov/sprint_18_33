# ПРОЕКТ: "В один клик". Персональные предложения для клиентов

Интернет-магазин «В один клик» продаёт разные товары: для детей, для дома, мелкую бытовую технику, косметику и даже продукты. Отчёт магазина за прошлый период показал, что активность покупателей начала снижаться. Привлекать новых клиентов уже не так эффективно: о магазине и так знает большая часть целевой аудитории. Возможный выход — удерживать активность постоянных клиентов. Сделать это можно с помощью персонализированных предложений.

«В один клик» — современная компания, поэтому её руководство не хочет принимать решения просто так — только на основе анализа данных и бизнес-моделирования. У компании есть небольшой отдел цифровых технологий, и вам предстоит побыть в роли стажёра в этом отделе. 

# Задача 
Разработать решение, которое позволит персонализировать предложения постоянным клиентам, чтобы увеличить их покупательскую активность.

# План Решения задачи:

<ol start="1">
    <li>Нужно промаркировать уровень финансовой активности постоянных покупателей. В компании принято выделять два
        уровня активности: «снизилась», если клиент стал покупать меньше товаров, и «прежний уровень».
    </li>
    <li>Нужно собрать данные по клиентам по следующим группам:
        <ul>
            <li>Признаки, которые описывают коммуникацию сотрудников компании с клиентом.</li>
            <li>Признаки, которые описывают продуктовое поведение покупателя. Например, какие товары покупает и как
                часто.
            </li>
            <li>Признаки, которые описывают покупательское поведение клиента. Например, сколько тратил в магазине.</li>
            <li>Признаки, которые описывают поведение покупателя на сайте. Например, как много страниц просматривает и
                сколько времени проводит на сайте.
            </li>
        </ul>
    </li>
</ol>

```python
def max_min(series):
    min_val = series.min()
    max_val = series.max()
    return max_val, min_val
```

```python
def upper_lower_borders(series, drop_nan):
    series = series.dropna() if drop_nan else series
    q1 = series.quantile(0.25)  # Нижний квартиль (Q1)
    q2 = series.quantile(0.50)  # Медиана (Q2)
    q3 = series.quantile(0.75)  # Верхний квартиль (Q3)

    # p25 = np.percentile(series, 25)    25-й процентиль с интерполяцией:
    # p75 = np.percentile(series,75)    75-й процентиль с интерполяцией:
    # difference = p75 - p25
    # iqr_value = stats.iqr(x)

    iqr = q3 - q1  # высчитываем размах
    lower_bound = q1 - 1.5 * iqr  # нижняя граница
    upper_bound = q3 + 1.5 * iqr  # верхняя граница
    return lower_bound, upper_bound
```

```python
class Plotter:
    def __init__(self, Series: pd.Series):
        self.Series = Series.dropna()  # очищаем от NaN для безопасности
        self.ax = None
        self.fig = None
        self.figsize = (10, 5)

    def set_figsize(self, figsize=(15, 5)):
        self.figsize = figsize
        return self

    def set_subplots(self, subplots_title, cols):
        self.fig, self.ax = plt.subplots(nrows=1, ncols=cols, figsize=self.figsize)
        self.fig.suptitle(subplots_title, fontsize=16)
        return self

    def set_title_ylabel_xlabel(self, title, ylabel, xlabel, index):
        if self.ax is not None:
            self.ax[index].set_title(title)
            self.ax[index].set_ylabel(ylabel)
            self.ax[index].set_xlabel(xlabel)
        return self

    def display_bar(self, bin=100, x_range=None, legend=None):
        try:
            sns.histplot(self.Series, kde=True, bins=bin, ax=self.ax[0], color='skyblue', edgecolor='black')
            if x_range:
                self.ax[0].set_xlim(x_range)
            if legend is not None:
                self.ax[0].legend([legend])
        except Exception as e:
            print("шкипер, пираты на горизонте:", e)
        return self

    def display_boxplot(self, horizontal: bool = False):
        try:
            sns.boxplot(x=self.Series if horizontal else None,
                        y=None if horizontal else self.Series,
                        ax=self.ax[1],
                        orient='h' if horizontal else 'v')
        except Exception as e:
            print("шкипер, пробоина в трюме:", e)
        return self

    # def display_Q_Q_Plot(self):
    #     st.probplot(self.Series, dist="norm", plot=self.ax[2])
    #     self.ax[2].set_title("Q-Q Plot")
    #     return self

    def show(self):
        plt.tight_layout()
        plt.show()


class BarPlotter:
    def __init__(self, series):
        self.series = series  # Series, а не DataFrame
        self.title = ""
        self.xlabel = ""
        self.ylabel = ""
        self.figsize = (10, 2)

    def set_title(self, title):
        self.title = title
        return self

    def set_xlabel(self, xlabel):
        self.xlabel = xlabel
        return self

    def set_ylabel(self, ylabel):
        self.ylabel = ylabel
        return self

    def set_figsize(self, width, height):
        self.figsize = (width, height)
        return self

    def plot(self):
        counts = self.series.value_counts().sort_values(ascending=False)
        plt.figure(figsize=self.figsize)

        ax = sns.barplot(x=counts.values, y=counts.index)
        ax.set_xlabel(self.xlabel)
        ax.set_ylabel(self.ylabel)
        ax.bar_label(ax.containers[0], fontsize=10)

        plt.title(self.title, fontsize=15)
        plt.tight_layout()
        plt.show()
```

```python
def separate_values(val, df):
    phik_data = (
        df.query('покупательская_активность == @val')
        .drop(['id', 'покупательская_активность'], axis=1)
    )
    interval_cols = phik_data.select_dtypes(include='number').columns.tolist()
    return phik_matrix(phik_data, interval_cols=interval_cols, bins=100)
```

```python
def display_corr_input_features(phik_overview_category, threshold=0.75):
    cor_data = []
    cor_result = []
    seen = set()

    for col in phik_overview_category.columns:
        for i, val in enumerate(phik_overview_category[col]):
            ind = phik_overview_category.index[i]
            if col == ind: continue
            if val >= threshold:
                cor_data.append((col, ind, val))
                # print(col,ind,val, sep='\t')

    for a, b, corr in cor_data:
        pair = frozenset([a, b])
        if pair not in seen:
            seen.add(pair)
            cor_result.append((a, b, corr))

    # Вывод
    if len(cor_result) == 0:
        print("все окей, нет коррелирующих признаков")
    else:
        for row in cor_result:
            print(row[0], " и ", row[1], " = ", row[2], sep="\t")
```

```python
ohe_columns = ["разрешить_сообщать", "популярная_категория"]
ord_columns = ["тип_сервиса"]
```

```python
ohe_list = [
    (
        'simpleImputer_ohe',
        SimpleImputer(missing_values=np.nan, strategy='most_frequent')
    ),
    (
        'ohe',
        OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False)
    )
]
ord_list = [
    (
        'simpleImputer_before_ord',
        SimpleImputer(missing_values=np.nan, strategy='most_frequent')
    ),
    (
        'ord',
        OrdinalEncoder(
            categories=[
                ['стандарт', 'премиум']
            ],
            handle_unknown='use_encoded_value',
            # Аргумент handle_unknown используется в кодировщиках категориальных признаков, таких как OneHotEncoder и OrdinalEncoder из sklearn. Он определяет, что делать с неизвестными категориями, которые встретятся в новых (тестовых) данных, но не были в обучающей выборке.
            unknown_value=-1  #unknown_value=np.nan
        )
    ),
    (
        'simpleImputer_after_ord',
        SimpleImputer(missing_values=np.nan, strategy='most_frequent')
    )
]
ohe_pipe = Pipeline(ohe_list)
ord_pipe = Pipeline(ord_list)
```

```python
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns),
        ('ord', ord_pipe, ord_columns),
        ('num', StandardScaler(), num_columns)
    ],
    remainder='passthrough'
)
```

```python
pipe_final = Pipeline(
    [
        ('preprocessor', data_preprocessor),
        ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
    ]
)
```

**аргументация**

| Метрика       | Когда использовать                                                                 | Особенности                                                                 |
|--------------|----------------------------------------------------------------------------------|-----------------------------------------------------------------------------|
| **Accuracy** | Данные сбалансированы (примерно равное количество объектов в классах)            | Не подходит для несбалансированных данных                                   |
| **Recall**   | Важно минимизировать пропуски положительного класса (например, диагностика болезней) | Оптимизирует обнаружение True Positive, но может увеличить False Positive   |
| **Precision** | Критично избегать ложных срабатываний (например, спам-фильтры)                   | Жёсткий контроль за False Positive, но может пропускать часть True Positive |
| **F1-score** | Нужен баланс между Precision и Recall                                            | Гармоническое среднее, полезно при дисбалансе классов                      |
| **ROC-AUC**  | Оценка качества модели независимо от порога классификации                        | Устойчива к дисбалансу, показывает общее качество разделения классов        |

выходит, что самая подходящая метрика - это метрика ROC-AUC

ROC-AUC — это площадь под ROC-кривой. Она оценивает качество решений модели в наиболее общем виде, учитывая все возможные пороговые значения. Поэтому она не зависит от того, какой порог будет выбран, это самая общая оценка итогов работы модели.

```python
# 1. Получаем лучшую модель из GridSearchCV
best_model = grid_search.best_estimator_

# 2. Подготавливаем данные через пайплайн
# Получаем препроцессор из пайплайна
preprocessor = best_model.named_steps['preprocessor']

# Трансформируем данные
X_train_transformed = preprocessor.transform(X_train)
X_test_transformed = preprocessor.transform(X_test)

# 3. Получаем имена фичей после трансформации
try:
    # Для новых версий sklearn
    feature_names = preprocessor.get_feature_names_out()
except AttributeError:
    # Для старых версий - создаем вручную
    feature_names = []

    # OHE фичи
    if 'ohe' in preprocessor.named_transformers_:
        ohe_transformer = preprocessor.named_transformers_['ohe']
        if hasattr(ohe_transformer.named_steps['ohe'], 'get_feature_names_out'):
            ohe_features = ohe_transformer.named_steps['ohe'].get_feature_names_out(ohe_columns)
            feature_names.extend(ohe_features)
        else:
            feature_names.extend([f"{col}_ohe" for col in ohe_columns])

    # ORD фичи
    if 'ord' in preprocessor.named_transformers_:
        feature_names.extend(ord_columns)

    # Числовые фичи
    if 'num' in preprocessor.named_transformers_:
        feature_names.extend(num_columns)

    # Остаточные фичи
    for name, trans, cols in preprocessor.transformers_:
        if name == 'remainder' and trans == 'passthrough':
            remainder_features = X_train.columns[cols]
            feature_names.extend(remainder_features)
# ------------------------------------------------------------------------------------------------------------
# 4. Создаем explainer для логистической регрессии
logistic_model = best_model.named_steps['models']

explainer = shap.LinearExplainer(
    logistic_model,
    X_train_transformed,
    feature_names=feature_names
)

shap_values = explainer.shap_values(X_test_transformed)

# font_settings = {
#     'font.size': 7,  # Общий размер шрифта
#     'axes.labelsize': 8,  # Подписи осей
#     'xtick.labelsize': 7,  # Метки оси X
#     'ytick.labelsize': 7,  # Метки оси Y (названия признаков)
#     'figure.titlesize': 9  # Заголовок
# }
# plt.rcParams.update(font_settings)
plt.figure(figsize=(14, 10), dpi=100)
# plt.figure(figsize=(10, 6))

# А. Важность признаков (глобальная)
shap.summary_plot(shap_values, X_test_transformed, feature_names=feature_names, plot_type="bar"
                  # , max_display=5
                  )
plt.show()
```

Перспективными и потенциальными покупателями можно считать тех, кто проводит на сайте больше времени, чем в среднем, с точки зрения длительности и частоты посещений. Также к ним относятся пользователи, которые просматривают больше страниц, чем среднестатистический посетитель, в том числе те, у кого более половины покупок являются акционными.

```python
total_data['вероятность_снижения'] = grid_search.predict_proba(X)[:, 1]
total_data = pd.merge(total_data, money_df, on='id', how='inner')
total_data["прибыль"] = total_data["прибыль"] * 1000
total_data.head()
```

```python
prob_thresh = total_data['вероятность_снижения'].median()
# -----------------------------------------------------------------------------------
def create_second_trash(col,trggr="md"):    
    if trggr == "md":
        inner_val = total_data[col].median()
    else:
        inner_val = total_data[col].mean()
    
    return inner_val

def assign_segment(row,col,second_trash):
    if row['вероятность_снижения'] > prob_thresh:
        if row[col] > second_trash:
            return 'Рискованный и прибыльный'
        else:
            return 'Рискованный и неприбыльный'
    else:
        if row[col] > second_trash:
            return 'Низкий риск и прибыльный'
        else:
            return 'Низкий риск и неприбыльный'
```

```python
def deper_display_segment(col,md_mn="md"):
    new_col = f"сегмент_{col}"
    secondary_trash = create_second_trash(col,md_mn)
    total_data[new_col] = total_data.apply(lambda x: assign_segment(x, col,secondary_trash), axis=1)

    def display_segment():
        plt.figure(figsize=(10, 6))
        sns.scatterplot(
            data=total_data,
            x='вероятность_снижения',
            y=col,
            hue=new_col,
            palette='Set2'
        )
        
        # Добавим линии порогов
        plt.axvline(prob_thresh, color='red', linestyle='--', label='Порог вероятности')
        plt.axhline(secondary_trash, color='blue', linestyle='--', label=f'Порог {col}')
        
        plt.title(f'Сегментация клиентов: по вероятности снижения и {col}')
        plt.grid(True)
        plt.legend()
        plt.show()
    return display_segment()
```
